# Place 1

```python
from pathlib import Path

pathlist = Path('C:/Users/s/Downloads/ttgeowork_roi1/').glob('**/*.tif') 
#folder where the Sentinel images for "place 1" for the concerned time period is saved

msavi2_mean_roi1 = [] #array for storing mean MSAVI value for "Place 1" for all timestamps
msavi2_median_roi1 = [] #array for storing median MSAVI value for "Place 1" for all timestamps

for path in pathlist: #iterating over all files (images) in the folder
    path_in_str = str(path)
    with rasterio.open(path_in_str) as src: 
        band_red = src.read(1) #reading the first band "B4" which is the Red band
    with rasterio.open(path_in_str) as src:
        band_nir = src.read(2) #reading the second band "B8" which is the NIR band
        
    #calculating MSAVI
    np.seterr(divide='ignore', invalid='ignore')
    a = 2*band_nir.astype(float) + 1
    b = band_nir.astype(float) - band_red.astype(float)
    msavi2_roi1 = (a - np.sqrt(a**2 - 8*b))/2
    
    #updating the msavi arrays initialised above
    msavi2_mean_roi1.append(np.nanmean(msavi2_roi1))
    msavi2_median_roi1.append(np.nanmedian(msavi2_roi1))
```

```python
#Plotting the MSAVI values with time

plt.plot(list(range(1,26)), msavi2_mean_roi1, color ='orange',
         marker ='o', markersize = 12, 
         label ='Mean MSAVI')

plt.plot(list(range(1,26)), msavi2_median_roi1, color ='green',
         marker ='o', markersize = 12, 
         label ='Median MSAVI')

plt.xlabel('Time')
plt.ylabel('MSAVI')

plt.title('MSAVI_roi1 vs Time')

plt.legend()
plt.show()
```

1. the "place 2" MSAVI value remain very close to zero for good chunk of images extracted for the time period.
